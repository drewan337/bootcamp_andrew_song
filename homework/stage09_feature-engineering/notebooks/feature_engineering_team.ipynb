```python
# Calculate 5-day and 20-day price momentum
df['price_momentum_5d'] = df['close'] / df['close'].shift(5) - 1
df['price_momentum_20d'] = df['close'] / df['close'].shift(20) - 1

# Create momentum signal (1 if 5d > 20d, else 0)
df['momentum_signal'] = (df['price_momentum_5d'] > df['price_momentum_20d']).astype(int)
```
